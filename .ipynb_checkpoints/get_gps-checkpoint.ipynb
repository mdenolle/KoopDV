# Download of GPS time series
This notebook collects GPS data given a lat-long box

```python
# import modules
import requests, zipfile, io, gzip, glob, os
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
%matplotlib inline
```

Collect in

```python

gps_loc = pd.read_csv("gps_locations.csv")
# this will yield a database of GPS station lat long
print(gps_loc)
```

```python
# select the stations within a lat long of interest, or pick your station
stalist = "AB01"
# index of the station chosen
ista=gps_loc[gps_loc["station"]==stalist].index.values.astype('int64')[0]



# select within a lat long box
latmin = 34
latmax = 36
longmin = -122
longmax = -119
ik=gps_loc[ (gps_loc["latitude"]>=latmin )& (gps_loc["latitude"]<latmax ) & (gps_loc["longitude"]>=longmin )& (gps_loc["longitude"]<longmax )].index.values.astype('int64')
stalist = gps_loc["station"].iloc[ik].values.tolist()
print(stalist)
print(len(stalist))
```

This section will download the data

```python
# This section downloads the data

# initialize nested lists
df=[]
for ista,sta in enumerate(stalist):
  print("Downloading and processing " + sta )
  file_url="http://geodesy.unr.edu/gps_timeseries/tenv/IGS14/"+ sta + ".tenv"
  r = requests.get(file_url).text.splitlines()  # download, read text, split lines into a list
  ue=[];un=[];uv=[];se=[];sn=[];sv=[];date=[];date_year=[]
  for iday in r:  # this loops through the days of data
    crap=iday.split()
    if len(crap)<10:
      continue
    date.append((crap[1]))
    date_year.append(float(crap[2]))
    ue.append(float(crap[6])*1000)
    un.append(float(crap[7])*1000)
    uv.append(float(crap[8])*1000)
#             # errors
    se.append(float(crap[10])*1000)
    sn.append(float(crap[11])*1000)
    sv.append(float(crap[12])*1000)

  # make dataframe
  crap={'station':sta,'date':date,'date_year':date_year,'east':ue,'north':un,'up':uv}
  if len(df)==0:
    df = pd.DataFrame(crap, columns = ['station', 'date','date_year','east','north','up'])
  else:
    df=pd.concat([df,pd.DataFrame(crap, columns = ['station', 'date','date_year','east','north','up'])])
print(df)


# plot time series
fig,ax=plt.subplots(3,1,figsize=(11,8),sharex=True)
for ista,sta in enumerate(stalist):
  ax[0].plot(df['date_year'][df['station']==sta],df['east'][df['station']==sta]);ax[0].grid(True);ax[0].set_ylabel('East (mm)')
  ax[1].plot(df['date_year'][df['station']==sta],df['north'][df['station']==sta]);ax[1].grid(True);ax[1].set_ylabel('North (mm)')
  ax[2].plot(df['date_year'][df['station']==sta],df['up'][df['station']==sta]);ax[2].grid(True);ax[2].set_ylabel('Up (mm)')
  ax[2].set_xlabel('Time (years)')
```

```python
# now let's find the trends and detrend the data.
from scipy import stats
Ve=[0]*len(stalist);Vn=[0]*len(stalist);Vu=[0]*len(stalist);
Vp=[0]*len(stalist);phi=[0]*len(stalist);
for ista,sta in enumerate(stalist):
  # linear regression such that: displacement = Velocity * time
  # velocity in the East component
  if len(df['east'][df['station']==sta])==0:
    continue
  Ve[ista], intercept, r_value, p_value, std_err = stats.linregress(df['date_year'][df['station']==sta],df['east'][df['station']==sta])
  Vn[ista], intercept, r_value, p_value, std_err = stats.linregress(df['date_year'][df['station']==sta],df['north'][df['station']==sta])
  Vu[ista], intercept, r_value, p_value, std_err = stats.linregress(df['date_year'][df['station']==sta],df['up'][df['station']==sta])

  # horizontal plate motion:
  Vp[ista]=(np.sqrt(Ve[ista]**2+Vn[ista]**2))
  phi[ista]=(90*np.arctan(Vn[ista]/Ve[ista]))
  print(sta,"overall plate motion there",Vp[ista],'mm/year')
  print(sta, "overall upward velocity",Vu[ista],'mm/year')
  print("strike",phi[ista])

```

```python
# plot detrended data 
from scipy import signal
fig,ax=plt.subplots(3,1,figsize=(11,8),sharex=True)
for ista,sta in enumerate(stalist):

  if len(df['east'][df['station']==sta])==0:
    continue

  Ve2 = signal.detrend(df['east'][df['station']==sta])
  Vn2 = signal.detrend(df['north'][df['station']==sta])
  Vu2 = df['up'][df['station']==sta] #signal.detrend(df['up'][df['station']==sta])

  ax[0].plot(df['date_year'][df['station']==sta],Ve2);ax[0].grid(True);ax[0].set_ylabel('East (mm)')
  ax[1].plot(df['date_year'][df['station']==sta],Vn2);ax[1].grid(True);ax[1].set_ylabel('North (mm)')
  ax[2].plot(df['date_year'][df['station']==sta],Vu2);ax[2].grid(True);ax[2].set_ylabel('Up (mm)')
  ax[2].set_xlabel('Time (years)')
```

```python
# make a map with the vectors
fig0=plt.figure(figsize=(8,6))
for ista,sta in enumerate(stalist):
  plt.quiver(gps_loc['longitude'][gps_loc['station']==sta],gps_loc['latitude'][gps_loc['station']==sta],Ve[ista]/100,Vn[ista]/100,scale=1,units='xy')
  plt.quiver(gps_loc['longitude'][gps_loc['station']==sta],gps_loc['latitude'][gps_loc['station']==sta],0,Vu[ista]/50,scale=1,color='r',units='xy')
plt.grid(True)
plt.axis('equal')
plt.xlim(-123,-118)
plt.ylim(32,39)
```
